```python
# Importar las bibliotecas necesarias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Cargar los datos de la elección
file_path = 'RESULTADOS_2024_CSV_V1.csv'  # Modificar con la ruta real al archivo CSV
datos_electorales = pd.read_csv(file_path)

# Visión general de los datos
print(datos_electorales.head())
print(datos_electorales.describe())

# Calcular el porcentaje de participación de votantes
datos_electorales['Porcentaje_Participacion'] = (datos_electorales['VOTOS_VALIDOS'] / datos_electorales['RE']) * 100

# Calcular el z-score de la participación
datos_electorales['Z_Score_Participacion'] = np.abs(stats.zscore(datos_electorales['Porcentaje_Participacion']))

# Definir un umbral para el z-score, valores por encima de 3 son considerados extremos
umbral = 3
anomalias = datos_electorales[datos_electorales['Z_Score_Participacion'] > umbral]

# Mostrar las anomalías detectadas basadas en la participación
print("Anomalías Detectadas Basadas en la Participación:")
print(anomalias[['EDO', 'MUN', 'PAR', 'CENTRO', 'Porcentaje_Participacion']])

# Visualización de las anomalías
plt.figure(figsize=(20, 6))
plt.scatter(datos_electorales.index, datos_electorales['Porcentaje_Participacion'], color='blue', label='Normal')
plt.scatter(anomalias.index, anomalias['Porcentaje_Participacion'], color='red', label='Anomalía')
plt.title('Anomalías en la Participación de Votantes')
plt.xlabel('Índice')
plt.ylabel('Porcentaje de Participación')
plt.legend()
plt.show()
```
